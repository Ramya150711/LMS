```python
train.isna().sum()
```

```python
train['Gender']=train['Gender'].fillna(train['Gender'].mode()[0])
```

```python
train['Married']=train['Married'].fillna(train['Married'].mode()[0])
```

```python
train['Dependents'].fillna(train['Dependents'].mode()[0],inplace=True)
```

```python
train['Self_Employed'].fillna(train['Self_Employed'].mode()[0],inplace=True)
```

```python
train['LoanAmount'].fillna(train['LoanAmount'].median(),inplace=True)
```

```python
train['Loan_Amount_Term'].fillna(train['Loan_Amount_Term'].mode(),inplace=True)
```

```python
train['Credit_History'].skew()
```

```python
train.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
feature_col=['Gender','Married','Dependents','Education','Self_Employed','Property_Area']
le=LabelEncoder()
for col in feature_col:
    train[col]=le.fit_transform(train[col])
```

```python
train.Loan_Status=train.Loan_Status.map({'Y':1,'N':0})
```

```python
train.head(3)
```

```python
train['total_income']=train['ApplicantIncome']+train['CoapplicantIncome']
```

```python
train.drop(columns=['ApplicantIncome','CoapplicantIncome'],inplace=True)
```

```python
train.head(3)
```

```python
rel_feat=['Gender','Married','Dependents','Education','Self_Employed','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area','total_income','Loan_Status']
```

```python
rel_feat_corr=train.corr()['Loan_Status'][['Gender','Married','Dependents','Education','Self_Employed','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area','total_income']]
```
